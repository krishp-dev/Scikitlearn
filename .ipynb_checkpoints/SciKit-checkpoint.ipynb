Column Transformers

```python
import numpy as np 
import pandas as pd
```

```python
x=pd.read_csv('covid_toy.csv')
```

```python
df=pd.DataFrame(x)
```

```python
df
```

```python
x=df.drop('has_covid',axis=1)
y=df['has_covid']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

Without Column Transfer

```python
from sklearn.impute import SimpleImputer
si=SimpleImputer()
x_train_fever=si.fit_transform(x_train[['fever']])
x_test_fever=si.transform(x_test[['fever']])
```

```python
from sklearn.preprocessing import OrdinalEncoder
ordinal=OrdinalEncoder(categories=[['Mild','Strong']])
x_train_cough=ordinal.fit_transform(x_train[['cough']])
x_test_cough=ordinal.transform(x_test[['cough']])
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder(drop='first',sparse_output=False)
```

```python
x_train_gender=ohe.fit_transform(x_train[['gender']])
x_test_gender=ohe.transform(x_test[['gender']])
x_train_city=ohe.fit_transform(x_train[['city']])
x_test_city=ohe.transform(x_test[['city']])
```

```python
x_train_transformed=np.concatenate((x_train_fever,x_train_gender,x_train_city),axis=1)
```

With Column Transformer

from sklearn.compose import ColumnTransformer

```python
tf=ColumnTransformer(transformers=[("tf1",SimpleImputer(),['fever']),("tf2",OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),("tf3",OneHotEncoder(drop='first',sparse_output=False),['gender'])])
```

```python
tf
```

```python
x_train_tf=tf.fit_transform(x_train)
x_test_tf=tf.transform(x_test)
```

Function Transfer

```python
import numpy as np
import pandas as pd

import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import FunctionTransformer
```

```python
df=pd.read_csv('Titanic-Dataset.csv',usecols=['Age','Fare','Survived'])
```

```python
df=pd.DataFrame(df)
```

```python
df
```

```python
df['Age']=df['Age'].fillna(df['Age'].mean())
```

```python
df['Fare']=df['Fare'].fillna(df['Fare'].mean())
```

```python
x=df.iloc[:,1:3]
y=df.iloc[:,0]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 8))

axs[0, 0].hist(x_train['Age'], bins=20, density=True, alpha=0.6, color='g')
axs[0, 0].set_title('Distribution - AGE')

stats.probplot(x_train['Age'], dist='norm', plot=axs[0, 1])
axs[0, 1].set_title('QQ PLOT - AGE')

axs[1, 0].hist(x_train['Fare'], bins=20, density=True, alpha=0.6, color='g')
axs[1, 0].set_title('Distribution - FARE')

stats.probplot(x_train['Fare'], dist='norm', plot=axs[1, 1])
axs[1, 1].set_title('QQ PLOT - FARE')

plt.tight_layout()
plt.show()
```

Training Models before Function Transfer

```python
clf=LogisticRegression()
clf2=DecisionTreeClassifier()
clf.fit(x_train,y_train)
clf2.fit(x_train,y_train)
y_pred1=clf.predict(x_test)
y_pred2=clf.predict(x_test)
print("Accuracy LR",accuracy_score(y_test,y_pred1))
print("Accuracy DT",accuracy_score(y_test,y_pred2))
```

Log Transformer

```python
trf=FunctionTransformer(func=np.log1p)
x_train_transfered=trf.fit_transform(x_train)
x_test_transfered=trf.transform(x_test)
```

```python
clf.fit(x_train_transfered,y_train)
clf2.fit(x_train_transfered,y_train)
y_pred=clf.predict(x_test_transfered)
y_pred2=clf.predict(x_test_transfered)
print("Accuracy LR",accuracy_score(y_test,y_pred1))
print("Accuracy DT",accuracy_score(y_test,y_pred2))
```

Reciprocal Transformer

```python
from sklearn.preprocessing import FunctionTransformer
import numpy as np

# Avoid divide-by-zero by adding a small constant (epsilon)
reciprocal_trf = FunctionTransformer(func=lambda x: 1 / (x + 1e-6))

x_train_transformed = reciprocal_trf.fit_transform(x_train)
x_test_transformed = reciprocal_trf.transform(x_test)
```

```python
from sklearn.metrics import accuracy_score

clf.fit(x_train_transformed, y_train)
clf2.fit(x_train_transformed, y_train)

y_pred1 = clf.predict(x_test_transformed)     # Classifier 1
y_pred2 = clf2.predict(x_test_transformed)    # Classifier 2

print("Accuracy LR", accuracy_score(y_test, y_pred1))
print("Accuracy DT", accuracy_score(y_test, y_pred2))
```

Square Transformer

```python
trf = FunctionTransformer(func=lambda x: np.power(x, 2))
x_train_transfered = trf.fit_transform(x_train)
x_test_transfered = trf.transform(x_test)
```

```python
clf.fit(x_train_transfered, y_train)
clf2.fit(x_train_transfered, y_train)
y_pred1 = clf.predict(x_test_transfered)
y_pred2 = clf2.predict(x_test_transfered)
print("Accuracy LR", accuracy_score(y_test, y_pred1))
print("Accuracy DT", accuracy_score(y_test, y_pred2))
```

 Square Root Transformation

```python
trf = FunctionTransformer(func=lambda x: np.sqrt(x + 1e-6))
x_train_transfered = trf.fit_transform(x_train)
x_test_transfered = trf.transform(x_test)
```

```python
clf.fit(x_train_transfered, y_train)
clf2.fit(x_train_transfered, y_train)
y_pred1 = clf.predict(x_test_transfered)
y_pred2 = clf2.predict(x_test_transfered)
print("Accuracy LR", accuracy_score(y_test, y_pred1))
print("Accuracy DT", accuracy_score(y_test, y_pred2))
```

 Box-Cox Transformation

```python
from sklearn.preprocessing import PowerTransformer
```

```python
# Box-Cox transformation only works for positive data, so we make sure to add a small value (e.g., 1e-6)
trf = PowerTransformer(method='box-cox')
x_train_transfered = trf.fit_transform(x_train + 1e-6)  # Adding a small constant to avoid 0 or negative
x_test_transfered = trf.transform(x_test + 1e-6)
```

```python
clf.fit(x_train_transfered, y_train)
clf2.fit(x_train_transfered, y_train)
y_pred1 = clf.predict(x_test_transfered)
y_pred2 = clf2.predict(x_test_transfered)
print("Accuracy LR", accuracy_score(y_test, y_pred1))
print("Accuracy DT", accuracy_score(y_test, y_pred2))
```

Yeo-Johnson Transformation

```python
trf = PowerTransformer(method='yeo-johnson')
x_train_transfered = trf.fit_transform(x_train)
x_test_transfered = trf.transform(x_test)
```

```python
clf.fit(x_train_transfered, y_train)
clf2.fit(x_train_transfered, y_train)
y_pred1 = clf.predict(x_test_transfered)
y_pred2 = clf2.predict(x_test_transfered)
print("Accuracy LR", accuracy_score(y_test, y_pred1))
print("Accuracy DT", accuracy_score(y_test, y_pred2))
```
